```python
import os
import json
import pandas as pd
```

```python
def counting(path):
    cnt = 0
    for filename in os.listdir(path):
        if os.path.isfile(os.path.join(path, filename)):
            cnt += 1
    return cnt
```

```python
target_path = "C:/Users/chowh/Desktop/컴과/프로젝트/UOSholar/AIHUB/용도별 목적 대화 데이터"
```

```python
target_path_list = os.listdir(target_path)
target_path_list
```

```python
total_data = 0
for i in range(len(target_path_list)):
    current_path = os.path.join(target_path, target_path_list[i])
    try:
        # 각 하위 디렉토리의 파일 개수를 셉니다.
        cnt = counting(current_path)
        total_data += cnt
        print(f'{target_path_list[i]}의 데이터 개수 = {cnt}')
    except FileNotFoundError:
        print(f"경로를 찾을 수 없습니다: {current_path}")
    
print(f'\n총 데이터 개수 = {total_data}')
```

```python
purpose = []

try:
    target_path_list = os.listdir(target_path)
except FileNotFoundError:
    print(f"경로를 찾을 수 없습니다: {target_path}")
    target_path_list = []

for folder_name in target_path_list:
    # 1. 하위 디렉토리의 전체 경로를 만듭니다.
    current_folder_path = os.path.join(target_path, folder_name)
    
    if os.path.isdir(current_folder_path):
        try:
            files = os.listdir(current_folder_path)
        except FileNotFoundError:
            print(f"경로를 찾을 수 없습니다: {current_folder_path}")
            continue

        for file_name in files:
            # 2. JSON 파일의 전체 경로를 만듭니다.
            final_path = os.path.join(current_folder_path, file_name)
            
            # JSON 파일만 처리하도록 확장자 확인
            if final_path.endswith('.json'):
                try:
                    with open(final_path, encoding="UTF-8") as target_file:
                        target_data = json.loads(target_file.read())
                        
                        # 'annotations' 키가 있는지 확인하여 오류 방지
                        if 'annotations' in target_data['info'][0]:
                            lines = target_data['info'][0]['annotations']['lines']
                            for line in lines:
                                if 'norm_text' in line:
                                    # 'A. ' 또는 'B. '를 제거하고 텍스트 추가
                                    text = line['norm_text'][2:]
                                    purpose.append(text.strip())
                                    
                except (json.JSONDecodeError, KeyError, IndexError, FileNotFoundError) as e:
                    print(f"오류 발생: {final_path} - {e}")

# 최종 결과 확인
print(f"총 {len(purpose)}개의 대화 텍스트를 추출했습니다.")
```

```python
import os
import json

target_path = r"C:/Users/chowh/Desktop/컴과/프로젝트/UOSholar/AIHUB/주제별 텍스트 일상 대화 데이터"

purpose = []

# target_path 안의 모든 파일 확인
for fname in os.listdir(target_path):
    if fname.lower().endswith(".json"):   # json 파일만 처리
        fpath = os.path.join(target_path, fname)

        try:
            with open(fpath, encoding="utf-8") as f:
                data = json.load(f)

            # info[0] → annotations → lines 안에서 대화문 추출
            info_list = data.get("info", [])
            if info_list and "annotations" in info_list[0]:
                lines = info_list[0]["annotations"].get("lines", [])
                for line in lines:
                    text = line.get("norm_text") or line.get("text")
                    if text:
                        # 발화 앞의 "1 :" 같은 부분 제거
                        text = text.split(":", 1)[-1].strip()
                        purpose.append(text)

        except Exception as e:
            print(f"[에러 발생] {fname} → {e}")

print(f"총 {len(purpose)}개의 대화 텍스트를 추출했습니다.")
print(purpose[:10])  # 앞부분 일부만 확인
```
